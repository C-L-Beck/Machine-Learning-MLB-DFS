```python
from fuzzywuzzy import fuzz

def matchnames(df1,df2,p = 0):
    dfx = pd.DataFrame(index=df1.index, columns=df2.index)

    for i in dfx.index:
        for j in dfx.columns:
            vi = df1.get_value(i, 'Name')
            vj = df2.get_value(j, 'Name')
            dfx.set_value(
                i, j, fuzz.token_sort_ratio(vi, vj))


    threshold = dfx.max(1) > 80
    idxmax = dfx.astype(float).idxmax(1)
    
    if p == 1:
    
        STAT1 = 'W'
        STAT2 = 'L'
        STAT3 = 'SV'
        STAT4 = 'G'
        STAT5 = 'GS'
        STAT6 = 'IP'
        STAT7 = 'K/9'
        STAT8 = 'BB/9'
        STAT9 = 'HR/9'
        STAT10 = 'BABIP'
        STAT11 = 'LOB%'
        STAT12 = 'GB%'
        STAT13 = 'HR/FB'
        STAT14 = 'ERA'
        STAT15 = 'FIP'
        STAT16 = 'xFIP'
        STAT17 = 'WAR'
        STAT18 = 'SwStr%'
        STAT19 = 'K%'
        


        df1['NameX'] = np.where(threshold, df2.loc[idxmax, 'Name'].values, np.nan)
        df1[STAT1] = np.where(threshold, df2.loc[idxmax, STAT1].values, np.nan)
        df1[STAT2] = np.where(threshold, df2.loc[idxmax, STAT2].values, np.nan)
        df1[STAT3] = np.where(threshold, df2.loc[idxmax, STAT3].values, np.nan)
        df1[STAT4] = np.where(threshold, df2.loc[idxmax, STAT4].values, np.nan)
        df1[STAT5] = np.where(threshold, df2.loc[idxmax, STAT5].values, np.nan)
        df1[STAT6] = np.where(threshold, df2.loc[idxmax, STAT6].values, np.nan)
        df1[STAT7] = np.where(threshold, df2.loc[idxmax, STAT7].values, np.nan)
        df1[STAT8] = np.where(threshold, df2.loc[idxmax, STAT8].values, np.nan)
        df1[STAT9] = np.where(threshold, df2.loc[idxmax, STAT9].values, np.nan)
        df1[STAT10] = np.where(threshold, df2.loc[idxmax, STAT10].values, np.nan)
        df1[STAT11] = np.where(threshold, df2.loc[idxmax, STAT11].values, np.nan)
        df1[STAT12] = np.where(threshold, df2.loc[idxmax, STAT12].values, np.nan)
        df1[STAT13] = np.where(threshold, df2.loc[idxmax, STAT13].values, np.nan)
        df1[STAT14] = np.where(threshold, df2.loc[idxmax, STAT14].values, np.nan)
        df1[STAT15] = np.where(threshold, df2.loc[idxmax, STAT15].values, np.nan)
        df1[STAT16] = np.where(threshold, df2.loc[idxmax, STAT16].values, np.nan)
        df1[STAT17] = np.where(threshold, df2.loc[idxmax, STAT17].values, np.nan)
        df1[STAT18] = np.where(threshold, df2.loc[idxmax, STAT18].values, np.nan)
        df1[STAT19] = np.where(threshold, df2.loc[idxmax, STAT19].values, np.nan)
#         df1[STAT20] = np.where(threshold, df2.loc[idxmax, STAT20].values, np.nan)
        
    else:
        
        STAT1 = 'G'
        STAT2 = 'PA'
        STAT3 = 'HR'
        STAT4 = 'R'
        STAT5 = 'RBI'
        STAT6 = 'SB'
        STAT7 = 'BB%'
        STAT8 = 'K%'
        STAT9 = 'ISO'
        STAT10 = 'BABIP'
        STAT11 = 'AVG'
        STAT12 = 'OBP'
        STAT13 = 'SLG'
        STAT14 = 'wOBA'
        STAT15 = 'wRC+'
        STAT16 = 'BsR'
        STAT17 = 'Off'
        STAT18 = 'Def'
        STAT19 = 'OPS'
        


        df1['NameX'] = np.where(threshold, df2.loc[idxmax, 'Name'].values, np.nan)
        df1[STAT1] = np.where(threshold, df2.loc[idxmax, STAT1].values, np.nan)
        df1[STAT2] = np.where(threshold, df2.loc[idxmax, STAT2].values, np.nan)
        df1[STAT3] = np.where(threshold, df2.loc[idxmax, STAT3].values, np.nan)
        df1[STAT4] = np.where(threshold, df2.loc[idxmax, STAT4].values, np.nan)
        df1[STAT5] = np.where(threshold, df2.loc[idxmax, STAT5].values, np.nan)
        df1[STAT6] = np.where(threshold, df2.loc[idxmax, STAT6].values, np.nan)
        df1[STAT7] = np.where(threshold, df2.loc[idxmax, STAT7].values, np.nan)
        df1[STAT8] = np.where(threshold, df2.loc[idxmax, STAT8].values, np.nan)
        df1[STAT9] = np.where(threshold, df2.loc[idxmax, STAT9].values, np.nan)
        df1[STAT10] = np.where(threshold, df2.loc[idxmax, STAT10].values, np.nan)
        df1[STAT11] = np.where(threshold, df2.loc[idxmax, STAT11].values, np.nan)
        df1[STAT12] = np.where(threshold, df2.loc[idxmax, STAT12].values, np.nan)
        df1[STAT13] = np.where(threshold, df2.loc[idxmax, STAT13].values, np.nan)
        df1[STAT14] = np.where(threshold, df2.loc[idxmax, STAT14].values, np.nan)
        df1[STAT15] = np.where(threshold, df2.loc[idxmax, STAT15].values, np.nan)
        df1[STAT16] = np.where(threshold, df2.loc[idxmax, STAT16].values, np.nan)
        df1[STAT17] = np.where(threshold, df2.loc[idxmax, STAT17].values, np.nan)
        df1[STAT18] = np.where(threshold, df2.loc[idxmax, STAT18].values, np.nan)
        df1[STAT19] = np.where(threshold, df2.loc[idxmax, STAT19].values, np.nan)
        #df1[STAT20] = np.where(threshold, df2.loc[idxmax, STAT20].values, np.nan)
        
    return df1

        
```
