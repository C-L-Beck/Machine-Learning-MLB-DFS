```python
import pandas as pd
from IPython.display import display
import numpy as np

#display all of dataframe
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
```

```python
pathc = "C:\\DATAMLBDFS\\ML\\MLB_Model_Complete.xlsx"
features_h = pd.read_excel(pathc, sheet_name='Batters') #Batters
features_p = pd.read_excel(pathc, sheet_name='Pitchers') #Pitchers

month_day = '05_17'
features_pred_h = pd.read_excel("C:\\DATAMLBDFS\\"+ month_day + "\\" + month_day + "_MLB.xlsx", sheet_name='Batters')
features_pred_p = pd.read_excel("C:\\DATAMLBDFS\\"+ month_day +"\\" + month_day + "_MLB.xlsx", sheet_name='Pitchers')

path = "C:\\DATAMLBDFS\\"+ month_day +"\\" + month_day + "Prediction_MLB.xlsx"

# month_day2 = '04_15'
# pathpred = "C:\\DATAMLBDFS\\" + month_day2 + "\\" + month_day2 + "_MLB.xlsx"
# features_pred = pd.read_excel(pathpred, sheet_name='Batters') #Batters
# features_pred_p = pd.read_excel(pathpred, sheet_name='Pitchers') #Pitchers
```

```python
features_h = features_h[features_h.Price != 0]
features_p = features_p[features_p.Price != 0]

display(features_h.head(50))
display(features_p.head(50))

```

```python
features_h.dtypes
```

```python
features_h = features_h.replace(np.nan,0)
features_h['BB%'] = features_h['BB%'].replace({'%':''}, regex = True)
features_h['BB%'] = features_h['BB%'].astype(float)
features_h['K%'] = features_h['K%'].replace({'%':''}, regex = True)
features_h['K%'] = features_h['K%'].astype(float)
features_h['Chosen'] = features_h['Chosen'].apply(int)
display((features_h).head(5))

Name = features_h['Name']
features_h.drop(labels=['NameX'], axis=1,inplace = True)
features_h.drop(labels=['Name'], axis=1,inplace = True)
features_h.drop(labels=['Team'], axis=1,inplace = True)
features_h.drop(labels=['Pos'], axis=1,inplace = True)
display((features_h).head(5))

```

```python
#stats and shape

features_h.shape
features_h.describe()
```

```python
#plot
```

```python
#dummy encode
features_h = pd.get_dummies(features_h)
features_h.head(5)
```

```python
#prediction column, labels are what we want to predict
labels = np.array(features_h['Chosen'])

#remove prediction column from main features
features_h = features_h.drop('Chosen', axis = 1)

#Saving feature names for later
features_h_list = list(features_h.columns)

#numpy array conversion
features_h = np.array(features_h)
```

```python
#split data into training and testing sets

# Using Skicit-learn to split data into training and testing sets
from sklearn.model_selection import train_test_split

# Split the data into training and testing sets x = features y = labels 
train_features, test_features, train_labels, test_labels = train_test_split(features_h, labels, test_size = 0.25, random_state = 42)
```

```python
print('Training Features Shape:', train_features.shape)
print('Training Labels Shape:', train_labels.shape)
print('Testing Features Shape:', test_features.shape)
print('Testing Labels Shape:', test_labels.shape)
```

```python
# Import the model we are using
from sklearn.ensemble import RandomForestClassifier
# Instantiate model with 1000 decision trees
rf = RandomForestClassifier(n_estimators = 1000, random_state = 42) #1000
# Train the model on training data
rf.fit(train_features, train_labels);
```

```python
# Use the forest's predict method on the test data
predictions = rf.predict(test_features)
```

```python
print (train_labels)
print (rf.predict(train_features))
```

```python
from sklearn.metrics import accuracy_score
print ("Train Accuracy :: ", accuracy_score(train_labels, rf.predict(train_features)))
print ("Test Accuracy  :: ", accuracy_score(test_labels, predictions))
```
